### without constructor

```python
class Area:
    len = 5
    bth = 10
    
    def show(self):
        print(self.name)
        
    def calc(self):
        ans = self.len * self.bth
        return ans, self.name

a = Area()
a.name = "sgm"
a.show()
a.calc()
a.name = "pot"
a.show()
a.calc()
print(a.name)
```

### using constructor

```python
class Area2:
    def __init__(self, len, bth):
        self.l = len
        self.b = bth
        
    def calc(self):
        ans = self.l * self.b
        print(self.l)
        return ans
```

### without constructor example

```python
o1 = Area()
o1.len = 5
o1.bth = 5
res = o1.calc()
print(o1.len, o1.bth, res)
```

```python
o2 = Area2(5,5)
print(o2.len, o2.bth)
```

```python
class Pokemons:
    def __init__(self, name, type="normal", hp=10, atk=10, dfn=90):
        self.name = name
        self.type = type
        self.hp = hp
        self.atk = atk
        self.dfn = dfn

    def status(self):
        print("Pokemon Details")
        print(f'Name: {self.name}')
        print(f'type: {self.type}')
        print(f'hp: {self.hp}')
        print(f'atk: {self.atk}')
        print(f'dfn: {self.dfn}')

    def overall(self):
        return self.hp + self.atk + self.dfn 
    
    def is_powerfull(self):
        if self.overall() > 1000:
            return True
        else:
            return False 
        
Pekachu = Pokemons("Pekachu", "Electric", 300, 400, 200)
Pekachu.status()
print(f'It\'s overall is {Pekachu.overall()}')
Pekachu.is_powerfull()
```

```python
class calc:
    def run(self, num1, operator, num2):
        match operator:
            case "+":  return self.add(num1, num2)   
            case "-":  return self.sub(num1, num2)
            case "/":  return self.div(num1, num2)    
            case "*":  return self.mul(num1, num2)          
            
    def add(self, a, b): return a + b    
    def sub(self, a, b): return a - b
    def div(self, a, b): return a / b
    def mul(self, a, b): return a * b 
    
do = calc()

while True:
    a = int(input("Enter first number"))
    b = input("Enter operators from these: +, -, / ,*")
    c = int(input("Enter second number"))
    ans = do.run(a, b, c)
    
    print(ans)
     
    if input("Do you want to continue") != "y":
        break
```

```python
class Human:
    def __init__(self, name):
        self.name = name
        
    def show_info(self):
        print("Details")
        print(f'name is {self.name}')
        
alex = Human("Alex marcer")
alex.show_info()

# ALL proprerties of parent class(Super class) will be inherited to child class(sub class)
class SuperHuman(Human):
    def __init__(self, heroName, name):
        self.heroName = heroName
        super().__init__(name)
        
    def secreat_identity(self):
        print(f'Secrate identity of {self.heroName} is {self.name}')
        
hero1 = SuperHuman("Batman", "Bravo")
hero2 = SuperHuman("Spidenman", "Peter")
hero1.show_info()
hero1.secreat_identity()
```

```python
class SI:
    def __init__(self, p ,r, t):
        self.p = p
        self.r = r
        self.t = t
        
    def calc(self):
        ans = (self.p + self.t) * (self.r / 100)
        return self.p + ans
    
class CI(SI):
    def calc(self):
        a = self.p * (1 + self.r / 100) ** self.t
        return a
        
entry1 = SI(9000, 2.5, 5)
print(f'SI = {entry1.calc()}')
        
entry2 = CI(9000, 2.5, 5)
print(f'CI = {entry2.calc()}')

#print(entry2 > entry1)
```

```python
class Player:
    def __init__(self, name, score):
        self.name = name
        self.score = score
        
    def __gt__(self, other):
        return self.score > other.score
    
    def __lt__(self, other):
        return self.score < other.score
    
    def __add__(self, other):
        return self.score + other.score
    
    def __sub__(self, other):
        return self.score - other.score
    
    def __eq__(self, other):
        return self.score == other.score
    
    """def __eq__(self, value):
        if isinstance(p1, c1)"""
    
        
p1 = Player("Dalinar", 60)
p2 = Player("Kaladin", 70)
p3 = Player("SGM", 40)
p4 = Player("PGM", 10)
p5 = Player("KGM", 60)

print(p1.score > p2.score)    
print(p1>p2)
print(p1<p2)
print(p1+p2)
print(p1-p2)
print(p1==p2)
```
