```python
for ascii_code in map(ord, ["a", "b", "A", "B"]):
    print(ascii_code, end=", ")
```

```python
print(list(map(round, [12.335, 64.85, 53.41, 4.88888])))
```

```python
def man(why):
    for this in why:
        yield this[0]

animals = ["wolf", "tiger", "lion" , "elephant", "rat"]
fruits = ["apple", "banana", "grape", "mango", "tomato"]
print(list(map(lambda a: a[0], animals)))

print(list(man(animals)))

print(list(map(lambda a: a[:3], animals)))

```

```python
### first 3 letters of animals and last 3letters of fruits
#this runs only if both parameters are of same length
list(map(lambda a, b: a[:3] + b[-3:], animals, fruits))
```

## filter 
same as map, ut the difference this accepts only the boolean fromat in return


```python
list(filter(lambda name: 'a' in name, animals))
```

```python
list(filter(lambda name: len(name) > 5, animals))
```

```python
class Student:
    def __init__(self, name, marks=[]):
        self.name = name
        self.marks = marks
        
    def total_marks(self):
        return sum(self.marks)
    
    def __gt__(self, other):
        if isinstance(other, Student):
            return self.total_marks() > other.total_marks()
        elif isinstance(other, int):
            return self.total_marks() > other
        
    def __repr__(self):
        return f'Student({self.name}, {self.total_marks()})'
    
    
def is_passed(marks):
    return marks > 2000

students = [Student("SGM", [5,54,85,85]), Student("mns", [5,54,8234,5,85]), Student("da", [5,54,85,85]),
            Student("dwqe", [5,5,85]), Student("dqwSGM", [5,54,85, 8,85]), Student("SGaM", [5,5234,4,85,85])]

list(filter(is_passed, students))
```
