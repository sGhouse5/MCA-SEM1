## Area and Perimeter of a rectangle 

```python
class Rectangle:
    def __init__(self, length, bredth):
        self.length = length
        self.bredth = bredth
        
    def find_area(self):
        print(f'Area is {self.length * self.bredth}')
    
    def find_perimeter(self):
        print(f'Perimeter is {2 * (self.length + self.bredth)}')
    
    
rectangle1 = Rectangle(2, 3)
rectangle1.find_area()
rectangle1.find_perimeter()
```

## Student's Total and Average of marks

```python
from statistics import mean

class Student:
    def __init__(self, name, mark1, mark2, mark3, mark4):
        self.name = name
        self.mark1 = mark1
        self.mark2 = mark2
        self.mark3 = mark3
        self.mark4 = mark4
        
    def find_total(self):
        print(f'Total marks of the student {self.name} are {sum([self.mark1, self.mark2, self.mark3, self.mark4])}')
    
    def find_average(self):
        print(f'Average marks of the student {self.name} are {mean([self.mark1, self.mark2, self.mark3, self.mark4])}')
    
    
student1 = Student("Ghouse", 10, 20, 30, 40)
student1.find_total()
student1.find_average()
```

## Temperature Converter

```python
class Temperature:
    def __init__(self, celsius):
        self.celsius = celsius
        
    def convert_to_kelvin(self):
        print(f'{self.celsius}° C = {self.celsius + 273.15} K')
    
    def convert_to_Fahrenheit(self):
        print(f'{self.celsius}° C = {(self.celsius * 9/5) + 32}° F')
    
    
input1 = Temperature(0)
input1.convert_to_kelvin()
input1.convert_to_Fahrenheit()
```

## Shopping Cart

```python
class ShoppingCart:    
    def __init__(self):
        self.cart = []


    def add_items(self, item, price):
        adding_item = {"item": item, "price": price}
        self.cart.append(adding_item)
        print(f'The item {adding_item["item"]} is added to the cart')        
    
    
    def remove_items(self, item):
        if len(self.cart) == 0:
            return print("No items added yet to remove, please add items")
        
        for item_in_cart in self.cart:
            if item_in_cart["item"] == item:
                self.cart.remove(item_in_cart)
                print(f'The item {item} is removed from the cart')
                return
        print(f'The item {item} is not present in the cart')
    
    
    def pricing(self):
        if len(self.cart) == 0:
            return print("Your Cart is Empty!, please add items")
    
        total_amount = 0
        print("*" * 30)
        print("The items in your cart are:")
        print(f'Items    Cost')
        
        for item in self.cart:
            total_amount += item["price"]
            print(f'{item["item"]}    {item["price"]}')
            
        print("-" * 15)
        print(f'Total    {total_amount}')
        print("*" * 30)
        
        
        
customer1 = ShoppingCart()
customer1.add_items("bat", 10000)
customer1.add_items("ball", 70)
customer1.add_items("cap", 66)
customer1.remove_items("cap")
customer1.remove_items("cap")
customer1.pricing()

customer2 = ShoppingCart()
customer2.remove_items("hhh")
customer2.pricing()
```

## Employee salary

```python
class Employee:
    def __init__(self, name, id, salary):
        self.name = name
        self.id = id
        self.salary = salary
        
    def check_status(self):
        print(f'\n Employee name: {self.name}\n Employee id: {self.id}\n Employee salary: {self.salary}\n')
        
    def give_a_raise(self):
        self.salary =  self.salary * 1.10
        print(f'Salary of the employee {self.name} is incremented by 10%')
        
employee1 = Employee("Ghouse", 65, 10)
employee1.check_status()
employee1.give_a_raise()
employee1.check_status()
```
