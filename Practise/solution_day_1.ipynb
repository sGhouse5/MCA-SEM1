## Problem 4 — Count vowels and consonants

## Problem 9 — Fibonacci
review

```python
n = int(input("Enter a number to find it's nth fibonacci number: "))
a, b = 0, 1
result = 0 

if n == 0:
    print(a)
elif n == 1:
    print(b)
else:
    for i in range(0, n-2):
        result = a + b
        a = b
        b = result  
    print(result)
```

## Problem 24 — Armstrong check


```python
actual_num = input("Enter the number you want to check: ")
checked_num = 0

for i in actual_num:
    i = int(i) ** len(actual_num)
    checked_num += i
    
if int(actual_num) == checked_num:
    print("Hey, it's Armstrong!")
else:
    print("Not an Armstrong")
    
    
```

## Problem 32 — Transpose matrix


```python
matrix = [[1, 2, 3, 4], [11, 22, 33, 44]]
Trans_matrix = []

for num, i in enumerate(matrix[0], start = 0):
    temp_column = []
    temp_column.append(i)
    for j in matrix[1]:
        if matrix[1].index(j) == num:
            temp_column.append(j)
            Trans_matrix.append(temp_column)
            break 
        

print("Given Matrix")
for i in matrix:
    print(i)

print("\nIts Transpose")
for i in Trans_matrix:
    print(i)
    
```
