## Boolean indexing

```python
x = np.random.randint(1, 10, (4, 8))
x
```

```python
x>7 #boolean filter- a matrix of true or false with sam edimensions of x matrix
```

```python
x[x>3]  # using boolean filter to extract values that are true
```

```python
filter1 = x > 3
filter2 = x < 7
# extract using more than 1 filter
x[filter1 & filter2] # or x[(x>3) & (x<7)]
```

## Useful methods and functions

```python
# np.arange(strat, stop, step)

a = np.arange(-10, 10, 3)
print(a)
```

```python
a = np.arange(-10, 10, .3)
print(a)
print(a.size)
```

```python
# np.linspace(strat, stop, num-of-elements) => start and stop are inclusive

a3 = np.linspace(2, 5, 15)
print(a3)
```

```python
# np.reshape(shape)
print(a3.size)

print("3 rows, 5 columns")
print(a3.reshape((3, 5)))

print("5 rows, 3 columns")
print(a3.reshape((5, 3)))
```

```python
z = np.linspace(1, 100, 30, dtype=int).reshape((5, 3, 2))
print(z)
```

```python
print(z.flatten())
```

```python
q = np.arange(-10, 10, .25)
x = np.sin(q)
y = np.cos(q)

print(q)
print(x)
print(y)
```

```python
import matplotlib.pyplot as plt
plt.plot(q, x)
plt.plot(q, y)
plt.show()
```
